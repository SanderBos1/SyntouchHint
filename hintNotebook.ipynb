In de afgelopen periode zijn verschillende SynTouchers druk in de weer geweest met de beveiliging van hun waardevolle data. Tot hun grote schrik ontdekten ze echter dat er toch een data lek was! Hoe kon dit gebeuren? Had iemand zijn laptop misschien wel per ongeluk open laten staan voor een inbreker?

Gelukkig heeft een scherpe collega ontdekt dat de gestolen data nog steeds uit de handen van de inbreker kan worden bevrijd. De informatie is opgeslagen op een geheime locatie, toegankelijk voor de inbreker, maar hij heeft zelf ook te maken met een data lek! Onze slimme collega heeft deze kwetsbaarheid weten te kraken.

Om de geheime locatie te achterhalen, moet je goed kijken naar de versleutelde data, die bestaat uit een mix van cijfers en letters. Door de letters te ordenen waarbij het bijbehorende nummer een priemgetal is, onthul je een boodschap die de sleutel tot de oplossing biedt.

Hieronder vind je een voorbeeld om je op weg te helpen:
